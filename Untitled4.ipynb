```python
import hashlib

# Known malware hashes (Example)
malware_hashes = [
    "5e884898da28047151d0e56f8dc6292773603d0d6aabbdd62a11ef721d1542d8"
]

filename = input("Enter file name: ")

try:
    with open(filename, "rb") as file:
        data = file.read()

    file_hash = hashlib.sha256(data).hexdigest()

    print("SHA-256 Hash:", file_hash)

    if file_hash in malware_hashes:
        print("Malware Detected!")
    else:
        print("File is Safe")

except FileNotFoundError:
    print("File not found!")
```

```python
url = input("Enter URL: ")

score = 0

if "@" in url:
    score += 1

if "-" in url:
    score += 1

if len(url) > 75:
    score += 1

if "https://" not in url:
    score += 1

if score >= 2:
    print("Warning: Possible Phishing URL")
else:
    print("URL appears Safe")
```

```python
from collections import Counter

with open("logs.txt", "r") as file:
    logs = file.readlines()

ips = []

for ip in logs:
    ips.append(ip.strip())

count = Counter(ips)

print("Request Count")

for ip in count:
    print(ip, ":", count[ip])

print("\nPossible DoS Attack")

for ip in count:
    if count[ip] > 3:
        print(ip, "made", count[ip], "requests")
```

```python
from scapy.all import sniff

def capture(packet):
    print(packet.summary())

print("Capturing 10 packets...")

sniff(prn=capture, count=10)

print("Capture Completed")
```

```python
import socket

host = input("Enter IP Address: ")

ports = [21, 22, 23, 25, 53, 80, 110, 443]

print("\nScanning...\n")

for port in ports:

    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.settimeout(1)

    result = s.connect_ex((host, port))

    if result == 0:
        print("Port", port, "is OPEN")
    else:
        print("Port", port, "is CLOSED")

    s.close()

print("\nScan Completed")
```
